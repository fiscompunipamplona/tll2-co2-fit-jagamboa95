$1)$ En la siguiente grafica se puede observar que la tendencia de cada mes, sin importar el año, es la misma, a diferencia del promedio de CO$_2$, el cual se observa que en cada mes este va aumentando, por ende, cada año hay más y más promedio de CO$_2$ en el aire, esta relación y comportamiento entre la tendencia y el promedio se puede observar en el grafico representada mediante la línea azul, mientras la línea naranja representa una tendencia promedio del comportamiento de la gráfica azul, esta primera grafica muestra esta relación entre los años 1980 y 2019. 


```python
from numpy import loadtxt,array, arange, linspace, sin, cos
from pylab import plot, show, savefig, ylim, xlim, grid, xlabel, ylabel
import matplotlib.pyplot as plt
dataco2 = loadtxt("../wd_alexander/unidad1_IntroLinux/data/co2_mm_gl.txt",float)

plt.plot(dataco2[:,2], dataco2[:,3],label='Datos')
plt.plot(dataco2[:,2], dataco2[:,4], label='Aproximación')
plt.ylim(330, 415)
plt.xlim(1980,2020)
plt.xlabel("Tendencia")
plt.ylabel("Promedio")
plt.title('Promedio de CO$_2$ desde 1980 al 2019')
plt.grid("on")
plt.legend()
show()
```

$2)$ Para el desarrollo de este segundo punto, lo que se busca es hallar una recta que defina la tendencia de incremento promedio a través de los años, para tres diferentes lapsos de tiempo, el primer grafico va a representar el tiempo entre los años 1980 y 2000, el segundo entre los años 2000 y 2019, y el tercer grafico para los años comprendidos entre 1980 y 2019.

Para la obtención de la ecuación de cada una de las rectar, utilizaremos el método de aproximación por mínimos cuadrados la cual nos permite hallar la pendiente y el intercepto con el eje Y de esta ecuación, para así poder graficar la recta promedio para cada una de las diferentes gráficas, las ecuaciones que me permiten hallar la pendiente y el intercepto con el eje Y son las siguientes:



\begin{equation}
m=\frac{\sum x \sum y - n\sum (xy)}{(\sum x)^2 -n\sum x^2 }
\end{equation}


\begin{equation}
b=\overline{y} -m\overline{x}
\end{equation}

De igual manera, obtendremos los valores de la regresión lineal, el cual nos permite observar que tan precisa es esta recta con respecto a los valores reales, sabiendo que entre más cercano a 1 este su valor, la aproximación es mucho más precisa y confiable, esta regresión es calculada mediante las siguientes ecuaciones:


\begin{equation}
R= \frac{\sigma_{xy}}{\sigma{x} \sigma{y}}
\end{equation}

\begin{equation}
\sigma_{x} = \sqrt{\frac{ \sum (x^2)}{n}-\overline{x}^2}
\end{equation}

\begin{equation}
\sigma_{y} = \sqrt{\frac{ \sum (y^2)}{n}-\overline{y}^2}
\end{equation}

\begin{equation}
\sigma_{xy} = \frac{\sum{xy}}{n} - \overline{x}\cdot\overline{y}
\end{equation}


```python
plt.plot(dataco2[0:240,2], dataco2[0:240,3],label='Datos')

n=len(dataco2[:240,2])
x=array(dataco2[:240,2])
y=array(dataco2[:240,3])
sumx=sum(x)
sumy=sum(y)
sumx2=sum(x*x)
sumy2=sum(y*y)
sumxy=sum(x*y)
promx=sumx/n
promy=sumy/n
sigmax=(sumx2/n - promx**2)**(1/2)
sigmay=(sumy2/n - promy**2)**(1/2)
sigmaxy=sumxy/n - promx*promy
R2= sigmaxy/(sigmax*sigmay)

m = (sumx*sumy - n*sumxy)/(sumx**2 - n*sumx2)
b=promy - m*promx 

plt.plot(x, m*x + b,label='Recta ajustada con minimos cuadrados')
plt.text(1981,365,"$R^2$=%.6f"%(R2))
plt.ylim(330, 375)
plt.xlim(1980,2000)
plt.xlabel("Tendencia")
plt.ylabel("Promedio")
plt.title('Promedio de CO$_2$ desde 1980 al 2000')
plt.grid("on")
plt.legend()
show()
```
